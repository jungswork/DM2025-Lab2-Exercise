## 1. Preprocessing Steps

#### (1) 資料整理為表格

```python
### Add the code related to the preprocessing steps in cells inside this section
import json
import pandas as pd

# 讀取 JSON 檔案
file_route = "./dm-lab-2-private-competition/final_posts.json"
with open(file_route, 'r', encoding='utf-8') as file:
    data = json.load(file)

# 建立表格前的資料處理: 將值由 JSON 轉換為 key - value list
data_list = []
for records in data:
    record = records["root"]["_source"]["post"]
    post_type = records["root"]["_type"]
    data_list.append({
                    "type": post_type,
                    "post_id": record["post_id"],
                    "text": record["text"],
                    "hashtags": record["hashtags"]
                    })
#print(data_list[:5]) check

# 建立表格
data_df = pd.DataFrame(data_list)
display(data_df.head())

# 將 test/train data 與 emotion data 根據相同的 id 加入表格內
# 先建立 test/train data 與 emotion data 表格
emotion_data_df = pd.read_csv("./dm-lab-2-private-competition/emotion.csv")
data_identification_df = pd.read_csv("./dm-lab-2-private-competition/data_identification.csv")

# 將 data_identification_df 切成 training data 及 test data
train_data_df = data_identification_df[data_identification_df["split"] == "train"]
test_data_df = data_identification_df[data_identification_df["split"] == "test"]

# 使用 merge() 將相對應的 id 資料匯入
# 匯入 emotion data
data_df = data_df.merge(emotion_data_df,
                        left_on= "post_id", # 左邊表格比較值
                        right_on= "id",     # 右邊表格比較值
                        how= "left"         # 依據左邊的 id 排列
                        )

# 匯入 training data
data_df = data_df.merge(train_data_df,
                        left_on= "post_id", 
                        right_on= "id",     
                        how= "left"        
                        )
# 合併後樣式
print("合併後樣式")
display(data_df.head())

# 處理 id_x, id_y
data_df = data_df.drop(columns=["id_x","id_y"])

# 清除 test data id(即移除空值)
data_df = data_df.dropna(subset=["split"])

print("最終版本")
data_df.head()
```

#### (2)Data preparation: check missing value and duplicated data

```python
# deal with empty value
# check dataframe total look
import helpers.data_mining_helpers as dmh

# 感謝助教的 code 我會心懷感激的使用的
display(data_df.isnull().apply(lambda x: dmh.check_missing_values(x)))
missing_num = data_df.isnull().apply(lambda x: dmh.check_missing_values(x))["emotion"][1]

print(f"空值比值: {missing_num/len(data_df):.2f}") # 無空值不須處理

# check duplicated data
print(f'重複資料數量: {sum(data_df["text"].duplicated())}') # 無 duplicated data 不須處理
```

#### (3)Data preparation: 視覺化檢查是否有資料不平衡情形

```python
import matplotlib.pyplot as plt
# emotion list
emotions = data_df["emotion"].unique().tolist()

# 計算 emotion 數量
emotion_counts = data_df["emotion"].value_counts().reindex(emotions) # 照著 emotion list 順序排列(不然會由大到小)
print(emotion_counts)

# 繪圖
plt.bar(emotions, emotion_counts.values)
plt.title("distribution of emotion")
plt.xlabel("emotion")
plt.ylabel("numbers")
plt.show()
"""
有不平衡的狀況， joy 太多了，需要分層取樣
"""
```

#### (4)Data preparation: Sampling 分層抽樣，將 joy, anger 類別抽樣，避免資料過度不平衡

```python
# sampling
joy_sampling_data = data_df[data_df["emotion"] == "joy"].sample(n= 10000)
anger_sampling_data = data_df[data_df["emotion"] == "anger"].sample(n= 8000)

# 更新 data_df
data_df = data_df[~data_df["emotion"].isin(["joy", "anger"])] # 不包含這兩個 emotion 的 data

# 將 sample 值加進去
data_df = pd.concat([data_df, joy_sampling_data, anger_sampling_data], ignore_index= True)

# 打亂排序
data_df = data_df.sample(frac=1, random_state=42)

#check
display(data_df)
len(data_df[data_df["emotion"] == "anger"])
```
